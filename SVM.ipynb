Import Packages

```
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
```

Define Create Chessboard Data Function

```
def CreateData(M,N):
  X1 = []
  X2 = []
  Y = []
  for i in range(M):
    for j in range(M):
      tmp1 = np.random.random(int(N/(M**2))+1) + j
      X1 = np.append(X1, tmp1) 
      tmp2 = np.random.random(int(N/(M**2))+1) + i  
      X2 = np.append(X2, tmp2)
      l = np.zeros(int(N/(M**2))+1) if (i+j)%2 == 0 else np.ones(int(N/(M**2))+1)
      Y = np.append(Y, l)
      c = 'r' if (i+j)%2 == 0 else 'b'
      plt.scatter(tmp1,tmp2, color=c) 
  
  plt.show()
  Data = np.zeros((len(X1),0))  
  Data = np.c_[Data, X1]
  Data = np.c_[Data, X2]  
  return Data , Y
```

Call CreateData Function

```
N = 10000
X , Y = CreateData(8,N/2)
print('Train Data = ',X.shape,'\n')
Xt , Yt = CreateData(8,N/2)
print('Test Data = ',Xt.shape,'\n')
```

Train Data and Predict

```
model = SVC(kernel='rbf' , gamma=2)
model.fit(X , Y)
Yp = model.predict(Xt)
print('Accuracy Score = ', accuracy_score(Yt, Yp))
print('Support Vectors For Each Class = ', model.n_support_)

plot_decision_regions(Xt, Yt.astype(int), model, legend=2 )
plt.title('SVM Decision Region Boundary', size=16)
plt.show()
```
