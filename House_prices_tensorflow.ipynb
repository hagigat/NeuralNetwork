```
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras.layers import Dense
from sklearn import preprocessing
```

```
dataset = pd.read_csv('train.csv')
dataset.dropna()
print(dataset.head(2))
```

```
dataset = dataset.select_dtypes(exclude=['object'])
dataset.drop('Id',axis = 1, inplace = True)
dataset.fillna(0,inplace=True)
print('numerical features:', dataset.shape)
print("List of features:",list(dataset.columns))
```

```
y_train = dataset[["SalePrice"]]
dataset.drop('SalePrice',axis = 1, inplace = True)
y_train = np.array(y_train)
x_train = np.array(dataset)
```

```
print(x_train.shape)
```

```
scaler = preprocessing.StandardScaler()
x_train = scaler.fit_transform( x_train )
y_train = scaler.fit_transform( y_train )
print(x_train[0])
print(y_train[0])
```

```
model = tf.keras.models.Sequential([
        Dense(36, input_dim=36 ,activation="sigmoid"),
        Dense(18, activation="sigmoid"),
        Dense(9, activation="sigmoid"),
        Dense(1),
])
```

```
model.compile(optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.5),
              loss='mean_squared_error',
              metrics=['mse'])
```

```
model.fit(x_train, y_train, epochs=200, shuffle=True)
```

```
dataset = pd.read_csv('test.csv')
dataset = dataset.select_dtypes(exclude=['object'])
dataset.drop('Id',axis = 1, inplace = True)
dataset.fillna(0,inplace=True)
print('numerical features:', dataset.shape)
print("List of features:",list(dataset.columns))
x_test = np.array(dataset)
```

```
dataset = pd.read_csv('sample_submission.csv')
y_test = dataset[["SalePrice"]]
y_test = np.array(y_test)
```

```
scaler = preprocessing.StandardScaler()
x_test = scaler.fit_transform( x_test )
y_test = scaler.fit_transform( y_test )
```

```
model.evaluate(x_test, y_test)
```

```
Y = model.predict(x_test)
for i in range(len(Y)):
  print("Actual :", y_test[i],"\tPredict : ", Y[i][0])
  print('___________________________________________')
```
